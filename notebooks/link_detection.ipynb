```python
from bs4 import BeautifulSoup
import re

from canvasapi import Canvas

import parsing
import secure
```

```python
canvas = Canvas(secure.API_URL, secure.API_KEY)
course = canvas.get_course(4358)
# course = canvas.get_course(4653)
```

```python
A = [a for a in course.get_assignments()]

A[2].description
```

```python
for a in A:
    soup = BeautifulSoup(a.description) if a.description else BeautifulSoup('')
    for link in soup.findAll('a', attrs={'href': re.compile("^http")}):
        print(link.get('href'))
```
